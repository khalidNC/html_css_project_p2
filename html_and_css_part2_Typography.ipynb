---

Flash of unstyled text means, the system font will show frist(for a while till custom font appears) in case the custom font takes much time to download.

There is way to create a situation where browser become slow for testing the above scenario.
1. Open the dev tool
2. Go to Network tab
3. Disable cache, in the top nav.
4. And set slow 3g network.
5. Now refresh the page

And you may notice it takes few sec to show the custom font and meanwhile it show the web safe one. This is called 'Flash of unstyled text'.

Now the problem is we can not fully solve this issue but we can minimize this. There a couple of properties we can use in the stylesheet.

**font-display:** property can be use inside font-face rule to tell browser how shoule handle the scenario. The default value is 'auto'

>a. font-display: auto; here browser decides what to show till custom font is downloaded. But it differ browser to browser.

    Chrome - In current version of chrome it initially show the fallback font then swap to custon when it is avilable.

    Other browsers- initially hide the text then show the custom font. This is called 'Flash of invisible text'.

>b. font-display: swap: This show the fallback font first then the custom one when it is avibale. Problem with this is, if take longer time to download the custom one there will be font breaks situstion for the end users.

>c. font-display: fallback; it is something like, if it takes longer time to download then show the fallback font.

>d. font-display: block; It is hide the text till the custom font downloads. Bu the problem is if it takes longer time or it not downloads then the text will not show up.

>e. font-display: optional; This probably gives the best experence. With this we arte telling browser;

    1. This font is optional and nice to have.
    2. So like fallback, we givi ng browser a short window to download and use the font.
    3. If download does not happen, the browser will solve the font.
    4. Behind the scene it tries to download it and store in it's cache.
    5. So after the first page load brower take the font form it's cache.

    With this user will not notice significent layout shift. So the only problem with is the font will be differnt for a while on the first page loads.
```
  @font-face {
    font-family: 'opensans';
    src: url('/typography/css/fonts/open-sans/opensans-regular-webfont.woff2') format('woff2'),
        url('/typography/css/fonts/open-sans/opensans-regular-webfont.woff') format('woff');
    font-weight: normal;
    font-style: normal;
    font-display: optional;
  }
```

###<font color="blue">***Sizing Font:***</font>
---

Sizing Font mean the mange the font size of your text.

**font-size property:**
1. Using the font-size property we can sizing the font.
2. value of this property could be absolute(e.g 10px 20px etc) or could be relative(1rem 2 rem ect)
3. As a general rule we should not use the absolute value because the end result could vary divice to device.
4. So we sould use the relative uints these are; %, vw(view port width), vh, em and rem).
5. For our case, we can set this to 1 rem, which means 1 time of the font size of the root element that is html doc.
```
  body {
    margin: 10px;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
    /* 1rem means, 1 time of the font size of the root element(html) */
    font-size: 1rem;
  }
```
6. But the question is, what is the size of the root element? Well, the html element font size depends on the user settings. Most the users do not change the default font size in Chrome which is 16px.(If you go to Chrome preference and search for the font size you is i is recommanded medium which is 16px).
7. So if we set a rule for html in the stylesheet it comes like. But we don want to explicitly set this with px.
```
  html {
    font-size: 16px;
  }
```
8. So since we set the body font-size to 1rem that mean it is 16px.
9. But if we set this to 1.5rem then the calculation(1.5x16) in px comes little bit complex. So to make it easy we convert 10px in precentage which is 62.5% equivalent to 10px. So we can set html font-size rule is 62.5% which is 10px. Then for the body element we can multiple by any numbers and convert in rem.
```
  html {
    font-size: 62.5%;
  }
```
```
  body {
    margin: 10px;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
    /* 1.6rem means, 1.6 time of the font size of the root element(html) 10px = 16px */
    font-size: 1.6rem;
  }
```

**font-size in different devices:**
1. We are going to do moblie first approach.
2. we set body text size to 1rem
3. Now set font-size for the h1. Let's say it is 2rem.
```
  h1 {
    font-family: Roboto, Arial, Helvetica, sans-serif;
    font-size: 2rem;
  }
```

**Apply Law of Proximity:**
The law of proximity, a fundamental principle in design, states that objects that are close to each other are perceived as belonging together. Applying the law of proximity means arranging text elements with appropriate vertical spacing to visually group related content. By adjusting the spacing between paragraphs, headings, and other text elements, designers can create a clear hierarchy and improve readability. Consistent vertical spacing helps users navigate the content more easily, enhancing the overall user experience. Utilizing CSS properties such as margin and padding, designers can control the vertical spacing between text elements to effectively apply the law of proximity and create visually cohesive layouts.

So we need maintain the low of proximity while writing multiple pragraphs and heading.

3. Since our body element font-size is 1rem, so the line-height should be 1.5rem
```
  body {
    margin: 10px;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
    font-size: 1rem;
    line-height: 1.5rem;
  }
```

```
  body {
    margin: 10px;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
    font-size: 1rem;
    line-height: 1.5;
  }
```

2. **text-decoration: underline;** this will make whole paragraph or heading underlined.
```
  .text-decoration p {
    text-decoration: underline;
  }
```

3. **text-decoration: line-through;** this will make line through the text. A usercase like price changed.

**text-transform:**
1. This is uesd to make text uppercase/lowrcase/capitalize.
2. **text-transform: uppercase;** with this youcan make the text all upppercase
3. **text-transform: lowercase;** Text can be all lowercase.
4. **text-transform: capitalize;** First letter of all the words will be cap.
```
  .text-transform {
    /* text-transform: uppercase */
    /* text-transform: lowercase; */
    text-transform: capitalize;
  }
```

**white-space:**
1. With this we can make the text trancated also called warpping.
2. **white-space: nowrap; this will make all the text in one line but no trancated.
3. **overflow:** Now snice we have width: 50ch for the text limit so we have a container that can take 50 charachters in 1 line. So we can keep 50ch rest of other will be tracated. but currently it is overfollow. And we know default overflow value is 'visible' so if we change this to 'hidden' then restof other will be trancated.
```
  .white-space p {
    white-space: nowrap;
    overflow: hidden;
  }
```
And now it is trancated but it is not clear to the reader thai is trancated. Better if there is a couple dot at the end.

  ![picture](https://drive.google.com/uc?id=1crxWxEPhXPpOZY3JfX5Y8HJrYBm6QYpP)

4. **text-overflow:** To do so we can use text-overflow property and set this to elipsis.
```
  .white-space p {
    white-space: nowrap;
    overflow: hidden;
    text-overflow: ellipsis;
  }
```

**column-count:**
1. This property is used to make the text in to multiple coulmns.
2. Can manage gap among the columns.
3. Make vertical line between the columns
```
  .white-space p:nth-of-type(2) {
    column-count: 2;
    column-gap: 2rem;
    column-rule: 3px dotted #999;
  }
```

**direction:**
1. This is used to make the text right to left direction.
2. we can make a single element right to left.
3. We can also make whole body content right to left.
```
  body {
    margin: 10px;
    font-family: system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;
    font-size: 1rem;
    line-height: 1.5;
    letter-spacing: 1px;
    word-spacing: 2px;
    width: 50ch;
    direction: rtl;
  }
```
