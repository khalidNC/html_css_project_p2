###<font color="blue">***CSS box model:***</font>
---

In HTML and CSS, the "box model" refers to the way elements on a web page are structured and displayed. It conceptualizes each element on a web page as a rectangular box with content, padding, borders, and margins.

Here's a breakdown of the components of the box model:

**Content:**
>This is the actual content of the HTML element, such as text, images, or other media.

**Padding:**
>Padding is the space between the content of the element and its border. It's used to create space around the content, inside the border.

**Border:**
>The border surrounds the padding and content of the element. It can be customized in terms of width, style, and color.

**Margin:**
>The margin is the space outside the border of the element. It creates space between the element and adjacent elements on the page.

When you set the dimensions (width and height) of an element in CSS, you're usually setting the dimensions of the content area only. The padding, border, and margin add to these dimensions.

For example, if you set an element's width to 200 pixels and its padding to 20 pixels, its total width would be 240 pixels (content width + left padding + right padding).

Understanding the box model is crucial for designing and laying out web pages effectively using HTML and CSS. It allows developers to control the spacing and dimensions of elements on the page precisely.

Here's a simple representation of the box model:
```
   ___________________________
  |  Margin                   |
  |   _____________________   |
  |  |  Border             |  |
  |  |   _______________   |  |
  |  |  |  Padding      |  |  |
  |  |  |   _________   |  |  |
  |  |  |  | Content |  |  |  |
  |  |  |  |_________|  |  |  |
  |  |  |_______________|  |  |
  |  |_____________________|  |
  |___________________________|
```

###<font color="blue">***Overflowing:***</font>
---

In HTML and CSS, the concept of overflowing occurs when the content within an element exceeds the dimensions set for that element. This typically happens when content is larger than its container, either in terms of width, height, or both.

Here's a brief overview of how overflowing is managed in HTML and CSS:

>**Overflow Property:** The overflow property in CSS is used to specify what happens if content overflows the content area of an element's box. It can take several values:

>>**visible:** Content is not clipped and may overflow the content box.

>>**hidden:** Content that overflows the box is hidden.

>>**scroll:** Scrollbars are added to the box to allow users to scroll and view overflowing content.

>>**auto:** Similar to scroll, but scrollbars are only added when necessary (i.e., when content overflows).

>**Overflow-x and Overflow-y:** These properties allow you to control overflow behavior independently for the horizontal (overflow-x) and vertical (overflow-y) dimensions.

>**Overflow Handling for Inline Elements:** By default, inline elements do not have an overflow. If content overflows an inline element, it may extend beyond the boundaries of its parent container. You can change this behavior by setting the inline element's display property to block or inline-block.

>**Overflow Handling for Flexbox and Grid Layouts:** In flexbox and grid layouts, the overflow behavior can be controlled at both the container and item level. Flex containers and grid containers have their own overflow properties, while individual flex items or grid items can have their own overflow behavior controlled using the same properties.

>**Overflow Handling for Positioned Elements:** When an element is positioned absolutely or fixedly, and its content overflows its containing block, it may extend beyond the viewport. In such cases, you can control the overflow behavior using overflow along with overflow-x and overflow-y properties.

###<font color="blue">***Measurement Units:***</font>
---

So far we measure the element with pixel. But there are others mesurement unites as well and they fall in to two catagories, Absolute and Relative.

**Absolute units** are always fixed, the value does not  change. Example are:
>>a. px(pixel)

>>b. pt(point)

>>c. in(Inch)

>>d. cm, mm and so on.

The cm and mm are used in priting not in web.

**Relative units** are someting else like:
>>a. % that relative to the size of container.

>>b. vw and vh which are relative to size of the viewport.

>>c. em and rem which are relative to the font size.

**Actions in codes:**
1. Add a div element defined a class=box
2. Add rules for the box in the stylesheet.
```
  .box {
      width: 150px;
      height: 150px;
      background-color: royalblue;
    }
```
We can see a box with the above dimention and color in the browser. And we resize the browser but the box size does noyt chaange. But when we build a responsive website the elements are need to be changed in size based on the browser's size. Let's describe few scenario one by one.

>>**Scenario 1:** Pixel value has applications on responsive web. Let's add a top border to the box. The top border is always 5px no matter the screen size is. So this is one application of pixel value.
```
  .box {
    width: 150px;
    height: 150px;
    background-color: royalblue;
    border-top: 5px solid gold;
  }
```

>>**Scenarion 2:** Relative value; let's change the width of the box to 50%. Means 50% of the parent element's value. What is the parrent element here? The box element is inside body element of body is the parent. So set rule body element(default 100% of the screen width) 300px. So in this case box width will 150px.
```
  body {
    margin: 10px;
    width: 300px;
  }
  .box {
    width: 50%;
    height: 150px;
    background-color: royalblue;
    border-top: 5px solid gold;
  }
```

So this relative value eventually converted to the pixel value by the browser.

Now remove the width property from the body element. So the default body element is 100% of the screen width since body is a block level element.
And this time we get the box width half of the screen size. And in this scenario if you resize the browser size the box size also resized in according to the browser size. Thais is the benefit of using relative unit.

3. **Viewport units:** Make the box height 100% of the height of the screen size.
```
  body {
    margin: 10px;
  }
  .box {
      width: 50%;
      height: 100%;
      background-color: royalblue;
      border-top: 5px solid gold;
    }
```
With this change our box disapears from the browser only the border presents. Because bydefault body element height is zero. So to fix this we can use the viewport units like vh(short form of viewport height).
```
  body {
    margin: 10px;
  }
  .box {
      width: 50%;
      height: 100vh;
      background-color: royalblue;
      border-top: 5px solid gold;
    }
```
Now the box height becomes full viewport height of the available vertical space of the screen and it is responsive.

And if we change the the width 50% to 50vw(viewport width) we get the same as before.
```
  body {
    margin: 10px;
  }
  .box {
      width: 50vw;
      height: 100vh;
      background-color: royalblue;
      border-top: 5px solid gold;
    }
```

4. **em and rem units:** These reletive units relative to the font size and we use in a situation, where we need to adjust the layout according to the font size. For example: If the font bigger we want the width to increase as well. Like if we set the width: 10em means the element size will be 10 times of the font size.

But what is the font size of this element? This element does not have the font size yet. So it will inherite the font size of it's parent element, which is body element. But the body element does not have the font size property. So it will inherites the font size of the html element. All the html element the base font size is 16px. So 10 times mean 10X16=160px will be the box width.
```
  box:nth-child(2) {
    width: 10em;
    height: 100px;
    background-color: red;
    border: none;
    margin-top: 10px;
  }
```

  ![picture](https://drive.google.com/uc?id=1mignpEMVljdvd98U4aIwoJdRNp7Jko6C)

Now if we set font-size: 20px the box width will be 20x10=200px
```
  .box:nth-child(2) {
    font-size: 20px;
    width: 10em;
    height: 100px;
    background-color: red;
    border: none;
    margin-top: 10px;
  }
```

**Drawback with em units:** if we do not explicitly set the font size then we need to find the root of the element to get the font-size which may bit complex. To overcome this complexity we can use another relative unit called 'rem'.

5. **rem unit:** rem unit is the xtimes of the root element, which is by default 16px so we just need remember only one number and no need to go by element's hierarchy.
```
  .box:nth-child(3) {
    /*means 10 times bigger than the root elelment's font size, 16px*/
    width: 10rem;
    height: 100px;
    background-color: red;
    border: none;
    margin-top: 10px;
  }
```

In HTML and CSS, the concept of "floating" is used to position elements within a container. When you float an element, you're essentially taking it out of the normal flow of the document and positioning it to the left or right within its containing element.

Here's a basic example of how floating works:

Html file:
```
  <!DOCTYPE html>
  <html lang="en">
    <head>
      <meta charset="UTF-8" />
      <meta name="viewport" content="width=device-width, initial-scale=1.0" />
      <title>Floating</title>
      <link rel="stylesheet" href="/normalize.css" />
      <link rel="stylesheet" href="/layout/css/floating.css" />
    </head>
    <body>
      <header><h1>Floating in HTML:</h1></header>
      <main>
        <article class="tweet">
          <div class="avator"></div>
          <p>
            In HTML and CSS, the concept of "floating" is used to position
            elements within a container. When you float an element, you're
            essentially taking it out of the normal flow of the document and
            positioning it to the left or right within its containing element.
          </p>
        </article>
      </main>
    </body>
  </html>
```

Flexible Box Layout(Flebox): It is used for laying out elements in one direction, in a row or column. It is a lot easier and powerful than floats.

Let's work with this in few examples:

4. Now define a set of rules in the stylesheet.
5. For caontainer, set border
```
  .container {
    border: 3px solid lightgray;
  }
```
6. And for boxes, we set width: 5rem, height: 5rem, Background: gold, and margin: 1rem

**Arrange these boxes horizentally in a row using flex, the common application of this, a navigation bar:**
1. First step is, go to the rule for the container in stylesheet and enable flex with display property to flex.
```
  .container {
    border: 3px solid lightgray;
    display: flex;
  }
```
Now tale a look, the boxes are laiedout horizentaly with a single line of code.

  ![picture](https://drive.google.com/uc?id=1M5dZyzPjHJsz-TK5HkaHKCyEARRx88I0)

1. We did have to use float
2. did not have to clear anything
3. did not have to manage with parent collapsing
4. With flex, we can layout a bunch of elements in one direction, in a column or row.

**Controll the flex direction using flex-direction property:**
1. inside the rules for the container, use flex-direction property.
2. And set this to (the default value is row, that's why it lays in a row) column and the boxes are laied out in a column.
```
  .container {
    border: 3px solid lightgray;
    display: flex;
    flex-direction: column;
  }
```

**Flex alignment:**

>***Axes in fex:*** To align elements we need understand the concept of axis in flex. In flex, we have 2 axises:
1.   Main (Primary Axis)
2.   Cross (Secondary Axis)

  >***Main axis:*** These axises are dependent on the direction. So if we set the direction to row, the main axis is going to be the hirizental axis. And the corss axis will be the vertical axis.

  >***Cross axis:*** If we set the direction to the column, the main axis will be the vertical axis and cross axis will be the horizental axis.

  ![picture](https://drive.google.com/uc?id=1tE_RMxWhU3qlbHXyizlEPFWTAHM3GIlB)

>***Align items inside container:*** With these axises we can easily align elements inside the container. To align items we need remenber 2 properties:
1.    justify-content: We use this to align items along the main axis.
2.    align-items: we use the to align items along the cross axis.

Most of the times these two properties are used to lign items.

**Application on Horizental Axis Using justify-content Property:**
**Align the boxes to the right:**
1. So we want to move the items to the end of the horizental axis.
2. This where we wil use justify-content property
3. So inside the container's rule add justify-content property ans defaul value is flex-start. Now set this to flex-end and the boxes are at the end of the horizental axis.
```
  .container {
    border: 3px solid lightgray;
    display: flex;
    flex-direction: row;
    justify-content: flex-end;
  }
```

**Align to center:**

**Distribute items along the horizental axis:**
There are 3 values for justiy-property to set:

>***a. space-evenly:*** this setting distributes the boxes having same space gap among them.

But at this point, in the frontend this does not affect because the container height is not big enough to push items. We need space at the botton of the container.

**Defining a Grid:**

4. and grid-tmeplate-column property to define the number and size of columns.

**Container in html file:**
```
  <body>
      <div class="container"></div>
  </body>
```

**Stylesheet rule:**

Let's create a grid of 3x2(3rows and 2columns) and size 100px.

```
  .container {
    display: grid;
    grid-template: repeat(3, 100px) / repeat(2, 100px);
    border: 3px solid gray;
  }
```

2. Make gap between rows and columns. There are 3 properties for gap. a) row-gap b) column-gap c) gap this for combining a and b.
```
  .container {
    display: grid;
    height: 90vh;
    grid-template: 100px auto 100px / 100px 30% 70%;
    row-gap: 10px;
    column-gap: 10px;
    border: 3px solid gray;
  }
```

Combine row and column gap by using gap property:
```
  .container {
    display: grid;
    height: 90vh;
    grid-template: 100px auto 100px / 100px 30% 70%;
    gap: 10x;
    border: 3px solid gray;
  }
```

**Placing items in grid:**
1. There three properties to placing elements inside grid. a)grid-row b)grid-column and c)grid-area(this is shorthabd for combining row and column)

2. With this property we can place any element wherever we want in row and column inside the grid. Fior example if we want to place the first box to entire first and second column then we can set the rule like;
>a. first in html define an extra class for the first box

```
  <div class="container1">
    <div class="con1 box-one">X</div>
    <div class="con1">Y</div>
    <div class="con1">Z</div>
    <div class="con1">ZZ</div>
  </div>
```

>b. Then set a new rule for the box-one as grid-column: 1 / 3;(which means this box place from line 1 to line 3) in stylesheet;
```
  .box-one {
    grid-column: 1 / 3;
  }
```
Now the frist box take place up to althrough the 2nd column.

  ![picture](https://drive.google.com/uc?id=1Zd90xV5vGc4IJD-uFjFvA9dZypPcI0oX)

>c. from first to last line set grid-column: 1 / -1;

>d. To move element through column set grid-column: 1 or 2 or 3 ...

>e. We can also use span keyword to place up to althrough the 2nd column, grid-column: 1 / span 2

>f. Move the first box to 2nd row:
```
  .box-one {
    grid-column: 1 / 3;
    grid-row: 2;
  }
```

**Placing items by defining spacific areas:**
1. We will define areas in our grid
2. Then place the items in the areas
3. There 2 properties to do this. a)grid-template-areas b)grid-area
4. grid-template-areas: define areas in container. Syntax is, grid-template-areas: "frist_row_column1 first_row_column2" "2nd_row_c1 2nd_row_c2" "3rd_row_c1 3rd_row_c2";
```
  .container1 {
    display: grid;
    height: 90vh;
    grid-template: 100px auto 100px / 30fr 70fr;
    gap: 10px;
    border: 3px solid gray;
    grid-template-areas: "header header" "sidebar main" "footer footer";
  }
```
5. grid-area: define individual area in container, now place the first box to header by using the grid-area property.
```
  .box-one {
    grid-column: 1 / 3;
    grid-area: header;
  }
```
So we do not need the position of row column any more. so remove the 'grid-column: 1 / 3;' and we will get same result.
```
  .box-one {
    grid-area: header;
  }
```

6. in same way define area for all the rows. Define class for last element in html then set rule for the element in stylesheet.
```
  <div class="container1">
    <div class="con1 box-one">X</div>
    <div class="con1">Y</div>
    <div class="con1">Z</div>
    <div class="con1 box-last">ZZ</div>
  </div>
```
Stylesheet:
```
  .box-last {
    grid-area: footer;
  }
```
And this automatically take full space of the container in last row or footer.

###<font color="blue">***Media Queries:***</font>
---

With media queries we can provide different styles for different devices depending on their features such as screen size, orientation and so on. So with this we can build website for mobile phone, tablets, and desktop computers. We call this website responsive website because they respond to various screen sizes.

We have two appraches to build responsive website.

>a. Build desktop version first then adjust layout for tablets and mobiles.

>b. or build the mobile version first then adjust the layout for others.

Defferent people have defferent choices but majority of the people choose mobile first. Becasue it easy to make a simple webiste for mobile and then adjust the the layout for desktop and tablets.

**Media Queries example:**

1. We have 2 paragraphs inside a boxes inside a container.
```
  <div class="container">
    <div class="box">
      <p>
        Lorem, ipsum dolor sit amet consectetur adipisicing elit. Deserunt
        neque impedit natus earum nobis veniam numquam sunt ducimus
        aspernatur non!
      </p>
    </div>
    <div class="box">
      <p>
        Lorem ipsum dolor sit amet consectetur adipisicing elit. Magnam
        dignissimos accusantium obcaecati voluptates voluptas eius!
      </p>
    </div>
  </div>
```
2. We are going to use flex to this container. By defaul flex direction is row. We will set direction to column for the mobile.

>a. Flex rule for container:
```
  .container {
    display: flex;
    flex-direction: column;
  }
```

>b. Rule for boxes:
```
  .box {
    background-color: gold;
    padding: 1rem;
  }
```

>c. Rule for the secound box using pseudo class:
```
  .box:nth-of-type(2) {
    background-color: royalblue;
  }
```
At this point the frontend looks like om mobile screen:

  ![picture](https://drive.google.com/uc?id=1tGhk06l_6BkrmFJbVzPoMVvVHfcsNqcO)

And on desktop it is not very good looking:

  ![picture](https://drive.google.com/uc?id=1jjAOy20_a4JRi_NupU0c-_EnGov4xLSz)

3. **Media Query:** Now we will use media queries and will set if screen size wider than let's say 500px then the derection will auto change to row and elements will show in a row.

>a. Find out the break point(viewport) manually resizing the browser, at that point you want to set media queries to put the boxes side by side in a row. Let's say the viewport width is 700px so at this point we will set media queries.

>b. to set media query, go to style sheet. the syntax is, @media 'query, can have multiple parts' {set_rule}

>c. Multiple parts are, a)First part is the type of the device, wich can be **screen** or **print**. @media screen

>d. The second part is the condition within the parenthesis and it start with 'and' oparator. @media screen and (condition: to check if the screen go behond 700px). Condition is written like; (min-width:700px).

>e. Now if the condition is ture then rules inside the curly braces will be apply. @media screen and (min-width: 700px) {set_rules_for_the_element}.
So in this case, we will set rule for the container, the felx direction to row if the condition is tule
```
  @media screen and (min-width: 700px) {
    .container {
      flex-direction: row;
    }
  }
```
Now the desktop where the screen width is more that the break point so there the boxes fit in a row as below.

  ![picture](https://drive.google.com/uc?id=1dz9-TGENgLf8-wR0fam04tBJ23qgPvJR)

>f. We can extend the query like this, we can set max-width to check the maximam screen size.
```
  @media screen and (min-width: 700px) and (max-width: 1100px){
    .container {
      flex-direction: row;
    }
  }
```
Now if the screen size in between 700 to 1100 px then the rule will be applied.

>g. we can also set a rule for a differnt screen size. let's say if my screen size more that 1000px width than the first box color change to purple.
```
  @media screen and (min-width: 1000px) {
    .container {
      flex-direction: row;
    }

    .box:nth-of-type(1) {
      background-color: purple;
    }
  }
```

>h. **print:** This media query to set rule for the printing only. Let say if we are going to print the content then we can set the media query for the font-size of the text.
```
  @media print {
    body {
      font-size: 12pt;
    }

    .box {
      padding: 0;
    }
  }
```
